# Fahrenheit to Celsius
In order to convert Fahrenheit to Celsius, we use the formula, °C = (°F - 32) × 5/9, in which the value of the temperature in Fahrenheit is placed and we get the value in Celsius.

Where,

C = Measure of temperature in degree Celsius (°C).

F = Measure of temperature in degree Fahrenheit (°F).

```python
# first we ask the user to enter the temperature in Fahrenheit and covert it to float
f = input("Write the temperature in Fahrenheit\n")
f = float(f) 
```

```python
# then Then we apply the law of conversion
c = (f - 32) * 5/9
```

```python
# finally print the result
print("the temperature in Celsius is ", c)
```
